```python
import pandas as pd

df=pd.read_csv('data/HugoSymbol.csv', sep=',',header=None)

print(df.values.size)
```

```python
#get first column
values_from_first_column = df[0].tolist()
print(values_from_first_column[0])

values_from_first_column[0:10]
```

```python
#add new column
df['Name'] = df[0].str.split(":").str[-1]
df
```

```python
#swap columns
df[[0,"Name"]] = df[["Name", 0]].where(df[0].str.contains('name'),"test")
df[[1,2]] = df[[2,1]].where(df[2].astype(str).str.startswith('pagerank'), df[[1,2]].values)

df
```

```python
#remove null values
df = df[df[2].notna()]
df
```

```python
df['EntrezGeneId'] = df[2].str.split(":").str[-1]

df = df[df['EntrezGeneId']!="NA"]
df
```

```python
df['pagerank'] = df[1].str.split(":").str[-1]
df
```

```python
from datetime import datetime
from elasticsearch import Elasticsearch
es = Elasticsearch(
    "http://localhost:9200",
    basic_auth=("elastic", "changeme")
)
```

```python
df = df.drop([0], axis=1)
df = df.drop([1], axis=1)
df = df.drop([2], axis=1)
df['Name'] = df['Name'].str.split(":").str[-1]
```

```python
for index, row in df.iterrows():
    doc = {'Name': row[0], 'EntrezGeneId': row[1], 'pagerank': row[2]}
    res = es.index(index="proteins", id=index, document=doc)
```

```python
res = es.get(index="proteins", id=66)
print(res)
```

```python
#res = es.search(index="proteins", query={"match_all": {}})
res = es.search(index="proteins", query={"match": {'Name': 'RER'}})
#res = es.search(index="proteins", query={"regexp": {'Name': 're.*'}})

print("Got %d Hits:" % res['hits']['total']['value'])
for hit in res['hits']['hits']:
    print(hit["_source"])
```
